```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
metadata_study = pd.merge(mouse_metadata,study_results, how="right", on=["Mouse ID"])
metadata_study.head()

# Display the data table for preview
print(metadata_study.head())
```

```python
# Checking the number of mice.
mice_count = len(pd.unique(metadata_study['Mouse ID']))
mice_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# Find the rows with duplicate Mouse ID and Timepoint
duplicate_rows = metadata_study[metadata_study.duplicated(subset=['Mouse ID', 'Timepoint'])]

# Extract unique Mouse IDs that have duplicates
duplicate_mouse_ids = duplicate_rows['Mouse ID'].unique()

print(f"Mouse IDs with duplicate timepoints: {', '.join(duplicate_mouse_ids)}")
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
# Get all data for the mouse IDs that have duplicate timepoints
all_data_for_duplicate_mice = metadata_study[metadata_study['Mouse ID'].isin(duplicate_mouse_ids)]

# Display the data
print(all_data_for_duplicate_mice)
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# Drop all data for the mouse IDs that have duplicate timepoints
clean_metadata_study = metadata_study[~metadata_study['Mouse ID'].isin(duplicate_mouse_ids)]

# Display the clean data
print(clean_metadata_study.head())

```

```python
# Checking the number of mice in the clean DataFrame.
mice_count = len(pd.unique(clean_metadata_study['Mouse ID']))
mice_count
```

## Summary Statistics

```python
clean_metadata_study.groupby("Drug Regimen").mean()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
means = clean_metadata_study.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
medians = clean_metadata_study.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variances = clean_metadata_study.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
stds = clean_metadata_study.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sems = clean_metadata_study.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
summary_stats_df= pd.DataFrame({"Mean Volumes": means, "Median Tumor Volumes": medians, "Variances in Tumor Volumes":variances,
                               "Standard Deviations of Volumes":stds, "SEM of Volumes":sems})
summary_stats_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
clean_metadata_study.describe()
# Using the aggregation method, produce the same summary statistics in a single line
summary_table = clean_metadata_study.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
summary_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
greatest_timepoint_df= clean_metadata_study.groupby("Mouse ID")[["Timepoint"]].max().reset_index()
greatest_timepoint_df
# Start by getting the last (greatest) timepoint for each mouse


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merge_df = pd.merge(greatest_timepoint_df,clean_metadata_study,how= "left",on=["Mouse ID", "Timepoint"])
merge_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list= []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments_list:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumors = merge_df.loc[merge_df["Drug Regimen"] == treatment]["Tumor Volume (mm3)"]
    
    # add subset 
    tumor_vol_list.append(tumors)
    
    # Determine outliers using upper and lower bounds
    q1 = tumors.quantile(0.25)
    q3 = tumors.quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = tumors.loc[(tumors < fence_low) | (tumors > fence_high)]
    print(f"{treatment}'s potential outliers: {df_out}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
flierprops = dict(marker='o', markerfacecolor='r', markersize=12,
                  linestyle='none', markeredgecolor='g')
plt.boxplot(tumor_vol_list,labels=treatments_list, flierprops=flierprops)

plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
cap_df = clean_metadata_study.loc[clean_metadata_study["Drug Regimen"] == "Capomulin"]
mouse_s185_data = cap_df[cap_df['Mouse ID'] == 's185']

mouse_s185_data.plot.line(x='Timepoint', y='Tumor Volume (mm3)') 
plt.title("Capomulin Treatment of Mouse s185")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Timepoint")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
cap_df = clean_metadata_study.loc[clean_metadata_study["Drug Regimen"] == "Capomulin"]
cap_agg_df= cap_df.groupby("Mouse ID")[["Tumor Volume (mm3)","Weight (g)"]].mean()
plt.scatter(x=cap_agg_df["Weight (g)"],y=cap_agg_df["Tumor Volume (mm3)"])
plt.ylabel("Average Tumor Volume (mm3)")
plt.xlabel("Weight (g)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
cap_df = clean_metadata_study.loc[clean_metadata_study["Drug Regimen"] == "Capomulin"]
cap_agg_df= cap_df.groupby("Mouse ID")[["Tumor Volume (mm3)","Weight (g)"]].mean()
x=cap_agg_df["Weight (g)"]
y=cap_agg_df["Tumor Volume (mm3)"]

plt.scatter(x,y)

slope, intercept, r, p, se = st.linregress(x, y)
y_=slope*x+intercept
plt.plot(x,y_,color="r")
print(f"The correlation between mouse weight and the average tumor volume is {round(r,2)}")

plt.ylabel("Average Tumor Volume (mm3)")
plt.xlabel("Weight (g)")

plt.show()
```
